```python


df.loc[0:2, 'DSP_Marks'] = [35.0, 38.0, 39.0]
#lambda arguments: expression (output)  anonymous(nameless) function without using def keyord
df['Pass/Fail'] = df['DSP_Marks'].apply(lambda x: 'Pass' if x >= 40 else 'Fail')
df
```

```python
plt.scatter(df[["DSP_Marks"]],df["Pass/Fail"])
```

```python
from sklearn.linear_model import LogisticRegression# Create a logistic regression model
model = LogisticRegression()


# Train the model on the training data
model.fit(df[["DSP_Marks"]], df["Pass/Fail"])
print(model.predict([[80]]))
print(model.predict_proba([[70]]))

```
