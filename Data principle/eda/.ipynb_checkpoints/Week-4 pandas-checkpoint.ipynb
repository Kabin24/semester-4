```python
#1 (average = mean)
df['Average marks']=0.0

df['Average marks']=df[["DSP_Marks","SSM_Marks","AP_Marks"]].mean(axis=1)
df ['Average marks']
```

```python

#2
noOfStudent = df[df["Average marks"] >= 80].shape[0]
print("No of student who have got above 80 average marks:",noOfStudent)
```

```python
df.agg(
    {
       "DSP_Marks": ["min", "max", "min"],
        "SSM_Marks": ["min", "max", "min"],

    }
)
```

```python

#Week 5 EDA part II Lab  (Visualizing Data with Matplotlib) : 
#Barchart
```

```python
groupdf=df.groupby("Gender")[["DSP_Marks","SSM_Marks","AP_Marks"]].mean()
```

```python
barchart =groupdf.plot(kind='bar')
barchart.set_title("Average Marks by Gender ")
barchart.set_ylabel("Average")
```

2)Histograms


```python
import matplotlib.pyplot as plt
plt.hist(df["DSP_Marks"],bins=[55,60,65,70,75,80,85,90,95],edgecolor="black")
import numpy as np
plt.yticks(np.arange(0,df.shape[0]))
plt.title("Histogram of DSP Marks")
plt.xlabel("DSP Marks")
plt.ylabel("No of Students")
plt.grid()
print(df.shape[0])
```

3)Visualizing proportions using Pie Chart

```python
df
```

```python
activities_counts = df["Extracurricular_Activities"].value_counts()
plt.figure(figsize=(5,5))
plt.pie(activities_counts, labels=activities_counts.index, autopct='%0.2f%%')
plt.title("Distribution of Extracurricular Activities")
```

#)Visualizing relationship between attendance and dsp marks
Scatter plot




```python
y= df["DSP_Marks"]
x = df["Attendance"]
plt.scatter(x, y, c='red', marker='o', edgecolors='black')
plt.xlabel("DSP Marks")
plt.ylabel("Attendance")
plt.title("Scatter Plot of DSP Marks vs. Attendance")
```
