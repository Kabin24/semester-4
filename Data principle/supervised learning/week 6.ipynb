```python
import pandas as pd
import matplotlib.pyplot as plt

import numpy as np
df=pd.read_csv('label data.csv')
df
```

```python
x = df["Study_Hours"]
y= df["DSP_Marks"]
plt.scatter(x, y, c='blue', marker='o', edgecolors='black')
plt.ylabel("DSP Marks")
plt.xlabel("Study Hrs")
plt.title("Scatter Plot of DSP Study vs. Study hrs")
```

```python
from sklearn.linear_model import LinearRegression


# Create a Linear Regression model
regr = LinearRegression()


# Calculate the average values for 'DSP_Marks' and 'Study_Hours' (ignoring NaN)
average_dsp_marks = df['DSP_Marks'].mean()
average_study_hours = df['Study_Hours'].mean()
# Fill NaN values with the calculated averages
df['DSP_Marks'].fillna(average_dsp_marks, inplace=True)
df['Study_Hours'].fillna(average_study_hours, inplace=True)


# Fit the model to the data
regr.fit(df[['Study_Hours']] ,df[['DSP_Marks']])
```

```python
predicted_DSP_Marks = regr.predict([[5.5]])
```

```python
yhat = regr.coef_ * df[["Study_Hours"]] + regr.intercept_
yhat
plt.plot(x,yhat,c='red')
```
